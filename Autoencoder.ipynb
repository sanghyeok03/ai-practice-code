```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

import pandas.api.types as ptypes

# 1. 데이터 불러오기
try:
    # Read each file separately
    df_info = pd.read_csv('cc_info.csv')
    df_transactions = pd.read_csv('transactions.csv')

    # --- Add code here to combine df_info and df_transactions into a single DataFrame 'df' ---
    # Example: If they can be concatenated (same columns)
    # df = pd.concat([df_info, df_transactions], ignore_index=True)
    # Example: If they can be merged on a common column (e.g., 'credit_card')
    # df = pd.merge(df_info, df_transactions, on='credit_card', how='inner')
    # Replace this placeholder with the actual logic to combine your data.
    # For now, let's assume 'df_transactions' is the primary data you want to work with for the rest of the script
    # If 'cc_info.csv' contains relevant features, merge or join appropriately.
    # As a temporary fix to get the script running with one file, let's use only transactions.csv
    df = df_transactions
    print("Data loaded successfully from transactions.csv.")


except FileNotFoundError as e:
    # Catching the specific error and printing which file was not found
    print(f"Error: {e}. Please check the file name and path.")
    exit() # Exit if files are not found
except Exception as e:
    # Catch any other potential errors during loading
    print(f"An unexpected error occurred during file loading: {e}")
    exit() # Exit on other errors too

# Check if df is created and not empty after loading and combining
if 'df' not in locals() or df.empty:
    print("Error: DataFrame 'df' was not created or is empty after loading and potential combining. Exiting.")
    exit()


# 2. 사용 컬럼 선택
# Note: Ensure the columns in 'cols' exist in your combined/selected 'df'
cols = ['credit_card', 'city', 'state', 'zipcode', 'credit_card_limit', 'date', 'transaction_dollar_amount', 'Long', 'Lat']

missing_cols = [col for col in cols if col not in df.columns]
if missing_cols:
    print(f"Warning: The following columns were not found in the DataFrame: {missing_cols}")
    cols = [col for col in cols if col in df.columns]
    if not cols:
        raise ValueError("Error: No valid columns found in the DataFrame after filtering. Exiting.")
    print(f"Proceeding with the following available columns: {cols}")

unique_cols = list(dict.fromkeys(cols))
if len(unique_cols) != len(cols):
    print("Warning: Duplicate column names found in the selected list. Using unique names.")
    cols = unique_cols


df_selected = df[cols].copy()

# 3. 범주형 변수 인코딩 (Label Encoding)
label_encoders = {}
for col in cols:
    # Check if the column exists in df_selected before processing
    if col in df_selected.columns:
        if ptypes.is_object_dtype(df_selected[col].dtype) or ptypes.is_categorical_dtype(df_selected[col].dtype):
            le = LabelEncoder()
            # Handle potential NaN values before fitting and transforming
            # fit_transform cannot handle NaN, so we'll fill with a placeholder
            # or handle before this step (e.g., in data loading/cleaning)
            # For now, a simple fillna:
            if df_selected[col].isnull().any():
                 print(f"Warning: Column '{col}' contains NaN values. Filling with 'Missing' for Label Encoding.")
                 df_selected[col] = df_selected[col].fillna('Missing')

            df_selected[col] = le.fit_transform(df_selected[col])
            label_encoders[col] = le
        else:
            print(f"Skipping Label Encoding for column '{col}' as it is not an object or categorical type.")
    else:
        print(f"Skipping Label Encoding for column '{col}' as it was not found in df_selected.")


# 4. 정규화
scaler = StandardScaler()

for col in df_selected.columns:
    # Ensure columns are numeric before attempting conversion (they should be after Label Encoding or if already numeric)
    # This loop is a bit redundant if label encoding already happened for object/categorical
    # However, it's safe for truly numeric columns initially selected.
    df_selected[col] = pd.to_numeric(df_selected[col], errors='coerce')


df_selected.dropna(inplace=True)


if df_selected.empty:
    print("Error: No data remaining after dropping rows with NaN values. Cannot proceed with scaling.")
    exit()

X_scaled = scaler.fit_transform(df_selected)

# 5. 학습 데이터 준비
if X_scaled.shape[0] < 2:
    print("Error: Not enough samples for train_test_split after processing. Cannot proceed with model training.")
    exit()

X_train, X_val = train_test_split(X_scaled, test_size=0.2, random_state=42)
X_train_tensor = torch.FloatTensor(X_train)
X_val_tensor = torch.FloatTensor(X_val)

if len(X_train_tensor) == 0:
    print("Error: Training data is empty after split. Cannot create DataLoader.")
    exit()

train_loader = torch.utils.data.DataLoader(X_train_tensor, batch_size=128, shuffle=True)

# 6. 오토인코더 정의
input_dim = X_scaled.shape[1]
if input_dim == 0:
     print("Error: Input dimension is zero. Cannot define Autoencoder.")
     exit()

class Autoencoder(nn.Module):
    def __init__(self, input_dim, encoding_dim=10):
        super(Autoencoder, self).__init__()
        if encoding_dim > input_dim:
            print(f"Warning: encoding_dim ({encoding_dim}) is greater than input_dim ({input_dim}). Setting encoding_dim to input_dim.")
            encoding_dim = input_dim
        # Ensure encoding_dim is at least 1 if input_dim is > 0
        if encoding_dim < 1 and input_dim > 0:
             encoding_dim = 1
             print(f"Warning: encoding_dim was less than 1, set to 1.")


        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, encoding_dim)
        )
        self.decoder = nn.Sequential(
            nn.Linear(encoding_dim, 64),
            nn.ReLU(),
            nn.Linear(64, input_dim)
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Check if input_dim is valid before creating the model
if input_dim > 0:
    model = Autoencoder(input_dim=input_dim)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
else:
    print("Error: Input dimension is zero. Cannot initialize Autoencoder model.")
    # Set model to None or exit, depending on desired behavior
    model = None
    criterion = None
    optimizer = None


# 7. 모델 학습
num_epochs = 50
# Check if model is initialized and train_loader has data before training
if model is not None and len(train_loader) > 0:
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for batch in train_loader:
            optimizer.zero_grad()
            outputs = model(batch)
            loss = criterion(outputs, batch)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {train_loss / len(train_loader):.4f}")
else:
    print("Warning: Model not initialized or train_loader is empty. Skipping model training.")


# 8. 인코딩된 데이터 추출
# Check if model is initialized and X_scaled has data before encoding
if model is not None and X_scaled.shape[0] > 0:
    model.eval()
    with torch.no_grad():
        encoded_data = model.encoder(torch.FloatTensor(X_scaled)).numpy()
else:
    print("Warning: Model not initialized or X_scaled is empty. Cannot extract encoded data.")
    encoded_data = np.array([])


# 9. DBSCAN 클러스터링

if encoded_data.shape[0] < 5:
    print("Warning: Not enough samples for DBSCAN with min_samples=5. Skipping DBSCAN clustering.")
    labels = np.zeros(encoded_data.shape[0]) # Create labels array of appropriate size, all 0
else:
    dbscan = DBSCAN(eps=0.5, min_samples=5)
    labels = dbscan.fit_predict(encoded_data)
    # Ensure labels length matches df_selected index length before indexing original df
    if len(labels) == len(df_selected.index):
        outliers_dbscan = df.loc[df_selected.index[labels == -1]]
        print(f"DBSCAN 이상치 탐지 수: {len(outliers_dbscan)}")
    else:
        print("Warning: Length of DBSCAN labels does not match df_selected index length. Cannot accurately identify outliers in original df.")
        outliers_dbscan = pd.DataFrame() # Initialize an empty DataFrame if mismatch


# 10. PCA 시각화
reduced = np.array([])
# Check if encoded_data has enough samples for PCA (at least n_components + 1 if n_components < min(n_samples, n_features))
# PCA requires at least 2 samples for n_components=2
if encoded_data.shape[0] > 1 and encoded_data.shape[1] >= 2: # Also check if there are at least 2 features
    try:
        reduced = PCA(n_components=2).fit_transform(encoded_data)
        # Check if PCA produced the expected output shape
        if reduced.shape[0] == encoded_data.shape[0] and reduced.shape[1] == 2:
            plt.figure(figsize=(8, 6))
            # Ensure labels length matches reduced data length for plotting
            if len(labels) == len(reduced):
                plt.scatter(reduced[:, 0], reduced[:, 1], c=labels, cmap='Spectral', s=5)
                plt.title("DBSCAN 결과 (PCA 2D)")
                plt.xlabel("PC1")
                plt.ylabel("PC2")
                plt.colorbar()
                plt.show()
            else:
                 print("Warning: Length of DBSCAN labels does not match reduced data length. Skipping PCA visualization.")
                 reduced = np.array([]) # Reset reduced if lengths mismatch
        else:
            print("Warning: PCA transformation did not produce expected shape. Skipping PCA visualization.")
            reduced = np.array([]) # Reset reduced if shape is wrong
    except ValueError as e:
        print(f"Warning: PCA failed: {e}. Skipping PCA visualization.")
        print("Check the shape of encoded_data:", encoded_data.shape)
        reduced = np.array([]) # Reset reduced if PCA fails
else:
    print("Warning: Not enough samples or features for PCA or encoded_data is empty. Skipping PCA visualization.")
    reduced = np.array([]) # Ensure reduced is empty if skipped


# 11. 재구성 오류 계산
with torch.no_grad():
    # Check if model is initialized and X_scaled has data before calculation
    if model is not None and X_scaled.shape[0] > 0:
        X_tensor = torch.FloatTensor(X_scaled)
        reconstructed = model(X_tensor)
        reconstruction_error = torch.mean((X_tensor - reconstructed) ** 2, dim=1).numpy()
    else:
        print("Warning: Model not initialized or X_scaled is empty. Cannot calculate reconstruction error.")
        reconstruction_error = np.array([])


# 12. 재구성 오류 시각화
if len(reconstruction_error) > 0:
    plt.figure(figsize=(8, 4))
    plt.hist(reconstruction_error, bins=50, color='skyblue', edgecolor='black')
    plt.title("재구성 오류 분포")
    plt.xlabel("MSE 재구성 오류")
    plt.ylabel("빈도")
    plt.grid(True)
    plt.show()
else:
    print("Warning: No reconstruction error data to visualize.")


# 13. 이상치 기준 설정
anomaly_mask = np.array([])
if len(reconstruction_error) > 0:
    # Calculate threshold only if reconstruction_error has enough data points (at least 2 for std)
    if len(reconstruction_error) > 1:
        threshold = np.mean(reconstruction_error) + 3 * np.std(reconstruction_error)
        print(f"이상치 탐지 임계값: {threshold:.4f}")

        anomaly_mask = reconstruction_error > threshold

        # Ensure anomaly_mask length matches df_selected index length before indexing original df
        if len(anomaly_mask) == len(df_selected.index):
            reconstruction_outliers = df.loc[df_selected.index[anomaly_mask]]
            print(f"재구성 오류 기준 이상치 탐지 수: {len(reconstruction_outliers)}")
        else:
             print("Warning: Length of anomaly_mask does not match df_selected index length. Cannot accurately identify outliers in original df.")
             reconstruction_outliers = pd.DataFrame() # Initialize an empty DataFrame if mismatch
    else:
        print("Warning: Not enough reconstruction error data points (need at least 2) to calculate threshold. Skipping outlier detection based on reconstruction error.")
        anomaly_mask = np.array([]) # Ensure anomaly_mask is empty


else:
    print("Warning: No reconstruction error data to analyze for outliers.")


# 14. 이상치 시각화
# Check if reduced data is available and lengths match anomaly_mask before plotting
if reduced.shape[0] > 0 and len(anomaly_mask) > 0 and len(anomaly_mask) == reduced.shape[0]:
    plt.figure(figsize=(8, 6))

    # Ensure boolean indexing works - ~anomaly_mask is boolean
    plt.scatter(reduced[~anomaly_mask, 0], reduced[~anomaly_mask, 1], c='blue', s=5, label='정상') # Use a fixed color for '정상'
    plt.scatter(reduced[anomaly_mask, 0], reduced[anomaly_mask, 1], color='red', s=8, label='이상치')
    plt.title("재구성 오류 기반 이상치 (PCA 2D)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("Warning: Not enough data points left after processing, data lengths mismatch, or no anomalies detected for visualization.")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import gc

# 1. 데이터 로딩 및 병합
try:
    df_info = pd.read_csv('cc_info.csv')
    df_transactions = pd.read_csv('transactions.csv')
    df = pd.merge(df_transactions, df_info, on='credit_card', how='inner')
    print("✅ 데이터 병합 완료. 총 샘플 수:", len(df))
except Exception as e:
    print(f"❌ 데이터 로딩 또는 병합 오류: {e}")
    raise e

# 2. 샘플 수 제한 (RAM 초과 방지)
if len(df) > 10000:
    df = df.sample(n=10000, random_state=42)
    print("✅ 10,000개로 샘플 제한")

# 3. 숫자형 컬럼 자동 탐색
numeric_df = df.select_dtypes(include=['int64', 'float64', 'int32', 'float32'])
numeric_cols = numeric_df.columns.tolist()

# 4. 필수 수치형 컬럼만 사용 (위도, 경도, 금액 등 포함되어 있으면 우선)
priority_cols = ['longitude', 'latitude', 'transaction_amount',
                 'credit_card_limit', 'zipcode', 'transaction_dollar_amount', 'Long', 'Lat']

# 실제 존재하는 우선 컬럼만 추림
used_cols = [col for col in priority_cols if col in numeric_cols]

# 그래도 없으면 그냥 숫자형 컬럼 전체 사용
if not used_cols:
    used_cols = numeric_cols[:5]  # 너무 많으면 RAM 과부하 우려 있으니 최대 5개

if not used_cols:
    raise ValueError("❌ 수치형 컬럼이 없습니다. 데이터 확인 필요")

print("✅ 사용되는 수치형 컬럼:", used_cols)

# 5. 결측치 제거 및 float32로 변환
df_filtered = df[used_cols].dropna().astype('float32')

# 6. 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_filtered)

# 7. k-distance plot (샘플 충분할 때만)
if X_scaled.shape[0] >= 5:
    neigh = NearestNeighbors(n_neighbors=5)
    nbrs = neigh.fit(X_scaled)
    distances, _ = nbrs.kneighbors(X_scaled)
    distances = np.sort(distances[:, 4])

    plt.figure(figsize=(8, 4))
    plt.plot(distances)
    plt.xlabel('Points sorted by distance')
    plt.ylabel('5-NN Distance')
    plt.title('k-distance plot (DBSCAN eps 선택 참고용)')
    plt.grid(True)
    plt.show()
else:
    print("⚠️ 샘플 수 부족으로 k-distance plot 생략")

# 8. DBSCAN 실행
eps_values = [0.5, 1.0]
min_samples_values = [5]

for eps in eps_values:
    for min_samples in min_samples_values:
        if min_samples > X_scaled.shape[0]:
            print(f"⚠️ Skipping eps={eps}, min_samples={min_samples}: 샘플 수 부족")
            continue

        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)
        n_outliers = np.sum(labels == -1)
        print(f"✅ eps={eps}, min_samples={min_samples} → 이상치 수: {n_outliers} / 총 샘플: {len(labels)}")

# 9. 가비지 컬렉션 호출 (RAM 회수)
gc.collect()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.metrics import mean_squared_error
from keras.models import Model
from keras.layers import Input, Dense
from keras import regularizers
import matplotlib.pyplot as plt
from scipy.spatial import distance

# 데이터 불러오기
transactions = pd.read_csv("transactions.csv")
cc_info = pd.read_csv("cc_info.csv")

# 병합
merged_df = pd.merge(transactions, cc_info, on="credit_card", how="inner")
print("병합 후 샘플 수:", len(merged_df))

# 수치형 컬럼 자동 선택
numeric_cols = merged_df.select_dtypes(include=[np.number]).columns.tolist()

# 이상치 감지에 사용할 수치형 컬럼만 사용 (이상값 있는 행 제거)
used_cols = [col for col in numeric_cols if merged_df[col].notnull().all()]
print("사용할 수치형 컬럼:", used_cols)

# 데이터 전처리
X = merged_df[used_cols].values
X = X[:10000]  # 데이터 수 너무 많을 경우 제한
X = StandardScaler().fit_transform(X.astype(np.float32))

# k-distance plot 시각화
from sklearn.neighbors import NearestNeighbors

k = 5
neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)

distances = np.sort(distances[:, k-1])
plt.figure(figsize=(8, 4))
plt.plot(distances)
plt.title(f"{k}-distance Plot")
plt.xlabel("Points sorted by distance")
plt.ylabel(f"{k}th Nearest Neighbor Distance")
plt.grid(True)
plt.show()

# DBSCAN eps 선택 및 이상치 탐지
print("\n📌 DBSCAN 이상치 탐지 결과:")
for eps in [0.5, 1.0, 1.5]:
    for min_samples in [5, 10]:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X)
        n_outliers = np.sum(labels == -1)
        print(f"▶ eps={eps}, min_samples={min_samples} ➜ 이상치 탐지 수: {n_outliers}")

# Autoencoder 재구성
encoding_dim = min(10, X.shape[1])  # 입력 차원 이상이면 자동 조정
input_dim = X.shape[1]
input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu',
                activity_regularizer=regularizers.l1(1e-5))(input_layer)
decoded = Dense(input_dim, activation='linear')(encoded)
autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='mse')

print("\n🧠 Autoencoder 학습 시작...")
autoencoder.fit(X, X, epochs=50, batch_size=64, shuffle=True, verbose=1)

# 재구성 오차 계산
reconstructions = autoencoder.predict(X)
mse = np.mean(np.power(X - reconstructions, 2), axis=1)
threshold = np.percentile(mse, 95)  # 상위 5%를 이상치로 간주
n_anomalies = np.sum(mse > threshold)
print(f"Autoencoder 이상치 탐지 수 (오차 기준): {n_anomalies} (Threshold: {threshold:.5f})")
```
